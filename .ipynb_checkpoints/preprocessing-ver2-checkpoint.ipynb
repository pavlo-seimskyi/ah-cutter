```python
import numpy as np
import librosa, librosa.display
import matplotlib.pyplot as plt
```

```python
file = "./audiofiles_chunks/0/1chunk0.wav" #0.1 seconds long
file2 = "./audiofiles/0/interview-cecilia-tham_recording-1_2020-05-25--t04-10-08pm--cecilia-1082.wav"
```

### Load audio file with Librosa

```python
signal, sample_rate = librosa.load(file, sr=16000)
```

**Display waveform**

```python
FIG_SIZE = (10, 5)
plt.figure(figsize=FIG_SIZE)
librosa.display.waveplot(signal, sample_rate, alpha=0.4)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("Waveform")
```

### STFT > spectrogram

```python
hop_length = 128 # in num. of samples
n_fft = 256 # window in num. of samples

# perform stft
stft = librosa.stft(signal, n_fft=n_fft, hop_length=hop_length)

# calculate abs values on complex numbers to get magnitude
spectrogram = np.abs(stft)

# display spectrogram
plt.figure(figsize=FIG_SIZE)
librosa.display.specshow(spectrogram, sr=sample_rate, hop_length=hop_length)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar()
plt.title("Spectrogram")
plt.show()
```

### Log spectrogram

```python
# apply logarithm to cast amplitude to Decibels

log_spectrogram = librosa.amplitude_to_db(spectrogram)

plt.figure(figsize=FIG_SIZE)
librosa.display.specshow(log_spectrogram, sr=sample_rate, hop_length=hop_length)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar(format="%+2.0f dB")
plt.title("Spectrogram (dB)")
plt.show()
```

### MFCCs - Mel Frequency Cepstral Coefficients

```python
# MFCCs

hop_length = 256 # in num. of samples
n_fft = 512 # window in num. of samples

MFCCs = librosa.feature.mfcc(signal, sample_rate, n_fft=n_fft, hop_length=hop_length, n_mfcc=10)

# display MFCCs
plt.figure(figsize=FIG_SIZE)
librosa.display.specshow(MFCCs, sr=sample_rate, hop_length=hop_length)
plt.xlabel("Time")
plt.ylabel("MFCC coefficients")
plt.colorbar()
plt.title("MFCCs")

# show plots
plt.show()
```
